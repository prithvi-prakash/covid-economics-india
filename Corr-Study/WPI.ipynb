```python
def getImpact(comm):
    pre_com = wpi[wpi["COMM_NAME"]==comm][pre_COI].to_numpy()
    post_com = wpi[wpi["COMM_NAME"]==comm][post_COI].to_numpy()
    print(pre_com)
    print(post_com)
    pre_corr, _ = pearsonr(list(pre_com[0]), list(pre["SI"]))
    post_corr, _ = pearsonr(list(post_com[0]), list(post["SI"]))
    print(f"Pre Pandemic {pre_corr}; Post Pandemic {post_corr}")
    direction = "Negative Shift" if pre_corr > post_corr else "Positive Shift"
    print(f"The direction of the Impact: {direction}.")
    print(f"Impact Magnitude {post_corr-pre_corr}")
```
